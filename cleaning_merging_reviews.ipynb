```python
import pandas as pd
import glob
import re

# ==========================================
# 1. MERGE THE 34 FILES
# ==========================================

# Use glob to find all files matching your naming pattern.
# Note: If your files are strictly Excel files, use "*.xlsx". 
# If they are CSVs, change the extension to "*.csv" and use pd.read_csv below.
file_pattern = r"c:\Users\emich\OneDrive - DVHE\Documents\ESILV\A4\S8\Information Retrieval and NLP\Project2\Traduction avis clients\Traduction avis clients\*.xlsx"
file_list = glob.glob(file_pattern)

print(f"Found {len(file_list)} files. Merging...")

# Read each file and store it in a list of DataFrames
dfs = []
for file in file_list:
    try:
        # Change to pd.read_csv(file) if your files are actually CSV format
        df = pd.read_excel(file) 
        dfs.append(df)
    except Exception as e:
        print(f"Error reading {file}: {e}")
```

```python
# Concatenate all 35 DataFrames into one massive DataFrame
df_merged = pd.concat(dfs, ignore_index=True)

# Drop any rows where the English review ('avis_en') might be missing
df_merged = df_merged.dropna(subset=['avis_en']).copy()
print(f"Total reviews after merging: {len(df_merged)}")
```

```python
import nltk
from nltk.corpus import stopwords

nltk.download('stopwords') 
stop_words = set(stopwords.words('english'))
```

```python
def clean_text(text):
    # 1. Convert text to string and lowercase it
    text = str(text).lower()
    # 2. Remove punctuation and numbers (keep only a-z and spaces)
    text = re.sub(r'[^a-z\s]', '', text)
    # 3. Tokenize (split into words)
    tokens = text.split()
    # 4. Remove stop words
    tokens = [word for word in tokens if word not in stop_words]
    # 5. Rejoin into a single clean string
    return " ".join(tokens)

print("Cleaning text...")
# Apply the cleaning function to create a new column
df_merged['clean_avis_en'] = df_merged['avis_en'].apply(clean_text)
```

```python
# ==========================================
# 3. SAVE THE FINAL DATASET
# ==========================================

# Save the final cleaned dataset to a CSV file for your NLP models
output_filename = "cleaned_reviews.csv"
df_merged.to_csv(output_filename, index=False)

print(f"Done! Cleaned dataset saved as '{output_filename}'.")
```
